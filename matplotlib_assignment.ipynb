```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

## Load the Dataset

```python
df = pd.read_csv("data/jumta.csv")
```

```python
df.info()
```

## Create a line plot showing the monthly revenue over the two-year period.
Customize the plot to make it publication-quality.

```python
# Data
x = df["Month"]
y = df["Revenue"]

plt.figure(figsize=(10,5))
# Line plot
plt.plot(x, y, label="Revenue", color="Purple")

# Add title and labels
plt.title('Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')

plt.xticks(rotation=45)
plt.legend()
# Show
plt.show()
```

## Create a bar plot showing the total number of orders per month.
Customize the plot to make it publication-quality.


```python
# Data
x = df["Month"]
y = df["Orders"]

plt.figure(figsize=(10,5))

plt.title("Orders Per Month", color="Green")
plt.xlabel("Month", color="Purple")
plt.ylabel("Orders", color="Purple")
plt.xticks(rotation=45)
plt.bar(x, y, label="Orders")
plt.grid(axis="y")
plt.legend()
plt.show()
```

## Create a scatter plot showing the relationship between the number of orders and the average order value.
Customize the plot to make it publication-quality.

```python
x = df["Orders"]
y = df["Average_Order_Value"]

plt.figure(figsize=(10, 5))
plt.title("Orders & Average Order value Relationship", color="Red")
plt.xlabel("Orders", color="Green")
plt.ylabel("Average Order Value", color="Green")
plt.scatter(x, y, label="Order Value")
plt.legend()
plt.show()
```

## Create a subplot containing all three visualizations created in steps 2-4.
Adjust the layout and size of the subplot to make it visually appealing.

```python
x1 = df["Month"]
y1 = df["Revenue"]

x2 = df["Month"]
y2 = df["Orders"]

x3 = df["Orders"]
y3 = df["Average_Order_Value"]

# Figure and axes
fig, (ax1, ax2, ax3) = plt.subplot()

# Plot
ax1.plot(x1, y1)
ax2.plot(x2, y2)
ax3.plot(x3, y3)

ax1.set_title('A Simple Line Plot')
ax1.set_xlabel('x-axis')
ax1.set_ylabel('y-axis')
```

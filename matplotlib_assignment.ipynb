```python
# Change the datatypes
new_R = []
for r in df["Revenue"]:
    d = r.split(",")
    revenue = d[0] + d[1]
    new_R.append(revenue)
new_R
df["Revenue"] = new_R

df["Revenue"] = df["Revenue"].astype("int64")

df.to_csv("data/new_jumta.csv", index=False)
```

 <span style="color: blue; font-size: 30px">Line Plot</span>

Create a line plot showing the monthly revenue over the two-year period.<br>
Customize the plot to make it publication-quality.

```python
# Data
x = df["Month"]
y = df["Revenue"]

plt.figure(figsize=(10,5))
# Line plot
plt.plot(x, y, label="Revenue", color="Purple")

# Add title and labels
plt.title('Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')

plt.xticks(rotation=45)
plt.legend()
# Show
plt.show()
```

<span style="color: blue; font-size: 30px">Bar Plot</span>

Create a bar plot showing the total number of orders per month.<br>
Customize the plot to make it publication-quality.

```python
# Data
x = df["Month"]
y = df["Orders"]

plt.figure(figsize=(10,5))

plt.title("Orders Per Month", color="Green")
plt.xlabel("Month", color="Purple")
plt.ylabel("Orders", color="Purple")
plt.xticks(rotation=45)
plt.bar(x, y, label="Orders")
plt.grid(axis="y")
plt.legend()
plt.show()
```

<span style="color: blue; font-size: 30px">Scatter Plot</span>

Create a scatter plot showing the relationship between the number of orders and the average order value.<br>
Customize the plot to make it publication-quality.

```python
x = df["Orders"]
y = df["Average_Order_Value"]

plt.figure(figsize=(10, 5))
plt.title("Orders & Average Order value Relationship", color="Red")
plt.xlabel("Orders", color="Green")
plt.ylabel("Average Order Value", color="Green")
plt.scatter(x, y, label="Order Value")
plt.legend()
plt.show()
```

<span style="color: blue; font-size: 30px">Subplot</span>

Create a subplot containing all three visualizations created in steps 2-4.<br>
Adjust the layout and size of the subplot to make it visually appealing.

```python
# Figure and axes
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(12,20))

x1 = df["Month"]
y1 = df["Revenue"]

x2 = df["Month"]
y2 = df["Orders"]

x3 = df["Orders"]
y3 = df["Average_Order_Value"]

# Revenue Per Month
ax1.plot(x1, y1, label="Monthly Revenue", marker="o", color="purple")
ax1.set_title("Revenue per Month", color="Red")
ax1.set_xlabel("Month", color="Green")
ax1.set_ylabel("Revenue", color="Green")
ax1.tick_params(axis="x", rotation=45)
ax1.grid(color='gray', linestyle='--', linewidth=0.5)
ax1.legend()

# Orders Per Month
ax2.bar(x2, y2, label="Monthly Orders", color="orange")
# ax2.set_title("Monthly Orders", color="red")
ax2.set_xlabel("Month", color="Purple")
ax2.set_ylabel("Orders", color="Purple")
ax2.grid(axis="y")
ax2.tick_params(axis="x", rotation=45)
ax2.legend()

# Number of orders and average order value
ax3.scatter(x3, y3, label="Number of Orders vs Average Order Value", color="red")
# ax3.set_title("Number of Orders vs Average Order Value", color="red")
ax3.set_xlabel("Orders", color="blue")
ax3.set_ylabel("Average Order Value", color="blue")
ax3.grid()
ax3.legend()

plt.show()
```
